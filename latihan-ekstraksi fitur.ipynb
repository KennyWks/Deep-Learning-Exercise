# Word2Vec

```python
# Import modul yang diperlukan
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize
import nltk
```

```python
# Download modul Punkt tokenizer dari NLTK
nltk.download('punkt_tab')
```

```python
# definisikan contoh data teks yang akan gunakan untuk melatih model Word2Vec
text_data = [
    'Saya suka makan bakso',
    'Bakso enak dan lezat',
    'Makanan favorit saya adalah nasi goreng',
    'Nasi goreng pedas adalah makanan favorit saya',
    'Saya suka makanan manis seperti es krim',
]
```

```python
#  Tokenisasi data teks menjadi daftar kata-kata
tokenized_data = [word_tokenize(sentence.lower()) for sentence in text_data]
print("Tokenized Data:", tokenized_data)
```

```python
# Membangun model Word2Vec menggunakan data teks yang sudah di-tokenisasi
model = Word2Vec(sentences=tokenized_data, vector_size=100, window=5, min_count=1, workers=4)
```

```python
#  Menggunakan vektor kata untuk kata-kata tertentu atau mencari kata-kata yang mirip dengan kata tertentu
word_vectors = model.wv
 
similar_words = word_vectors.most_similar('bakso', topn=3)
print("Kata-kata yang mirip dengan 'bakso':", similar_words)
 
vector = word_vectors['bakso']
print("Vektor untuk 'bakso':", vector)
```

# TF-IDF

```python
# import module
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# Some text in documents
documents_tf_idf = [
    "Saya suka makan bakso",
    "Bakso enak dan lezat",
    "Makanan favorit saya adalah nasi goreng",
    "Nasi goreng pedas adalah makanan favorit saya",
    "Saya suka makanan manis seperti es krim",
]
```

```python
# Inisialisasi objek TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer()
```

```python
# menghitung TF-IDF dari dokumen
tfidf_matrix = tfidf_vectorizer.fit_transform(documents_tf_idf)
```

```python
# melihat vocabulary (kata unik) yang dihasilkan oleh TfidfVectorizer
print("Vocabulary:", tfidf_vectorizer.vocabulary_)
```

```python
# melihat hasil dari TF-IDF matrix dalam bentuk arra
print("TF-IDF Matrix:")
print(tfidf_matrix.toarray())
```

# Bag of Words (BoW)

```python
# import module
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# Data teks berisi beberapa dokumen contoh yang akan diolah
documents_BoW = [
    "Ini adalah contoh dokumen pertama.",
    "Ini adalah dokumen kedua.",
    "Ini adalah dokumen ketiga.",
    "Ini adalah contoh contoh contoh."
]
```

```python
# Inisialisasi objek CountVectorizer 
vectorizer = CountVectorizer()
```

```python
# fitting dan transformasi pada data teks menggunakan CountVectorizer
bow_matrix = vectorizer.fit_transform(documents_BoW)
```

```python
# Matriks ini berisi jumlah kemunculan setiap kata dalam setiap dokumen
bow_matrix.toarray()
```

```python
# fitur (kata-kata) yang dihasilkan oleh CountVectorizer
features = vectorizer.get_feature_names_out()
```

```python
print("Matriks BoW:")
print(bow_matrix.toarray())
 
print("\nDaftar Fitur:")
print(features)
```

# N-gram

```python
# import module
from nltk.util import ngrams
```

```python
# Data teks berisi beberapa kalimat contoh yang akan diolah
sentences = [
    "Saya suka makan bakso enak di warung dekat rumah.",
    "Nasi goreng adalah salah satu makanan favorit saya.",
    "Es krim coklat sangat lezat dan menyegarkan.",
    "Saat hari hujan, saya suka minum teh hangat.",
    "Pemandangan pegunungan di pagi hari sangat indah.",
    "Bola basket adalah olahraga favorit saya sejak kecil."
]
```

```python
for sentence in sentences:
    # Bagi kalimat menjadi kata-kata individu
    words = sentence.split()
 
    # Buat 1-gram, 2-gram, dan 3-gram dari kata-kata tersebut
    unigrams = list(ngrams(words, 1))
    bigrams = list(ngrams(words, 2))
    trigrams = list(ngrams(words, 3))
 
    # Cetak hasil ekstraksi N-gram ke layar
    print("\nKalimat:", sentence)
 
    print("1-gram:")
    for gram in unigrams:
        print(gram)
 
    print("\n2-gram:")
    for gram in bigrams:
        print(gram)
 
    print("\n3-gram:")
    for gram in trigrams:
        print(gram)
```
